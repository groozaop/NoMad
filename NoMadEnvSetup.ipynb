```python
import requests
import concurrent.futures
import os
import threading # Import threading for the Lock and the counter/limit logic

# --- Configuration ---
PROXY_LIMIT = 100 # The target number of working proxies to find
lock = threading.Lock()
alive_count = 0

def check_proxy(proxy):
    """Checks if a single proxy is alive against Google."""
    url = "http://www.google.com" # URL to test the proxy against
    proxies = {
        "http": f"http://{proxy}",
        "https": f"http://{proxy}",
    }
    # Check if the limit has already been hit before starting the request
    global alive_count
    if alive_count >= PROXY_LIMIT:
        return proxy, False # Return False to immediately stop processing

    try:
        response = requests.get(url, proxies=proxies, timeout=5)
        if response.status_code == 200:
            return proxy, True
    except:
        return proxy, False
    return proxy, False

def load_proxies(file_path):
    """Loads proxies from a file."""
    try:
        with open(file_path, 'r') as file:
            proxies = file.readlines()
        return [proxy.strip() for proxy in proxies]
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return []

def write_alive_proxy(proxy, save_path):
    """Writes a single alive proxy to the file, safely using a lock."""
    global alive_count
    with lock:
        # Check limit again just before writing/incrementing (double-check lock)
        if alive_count < PROXY_LIMIT:
            with open(save_path, 'a') as file: # Use 'a' for append mode
                file.write(f"{proxy}\n")
            alive_count += 1
            print(f"Proxy {proxy} is alive. Count: {alive_count}/{PROXY_LIMIT}")
            return True # Successfully wrote and incremented
        return False # Limit already met

def main():
    global alive_count

    file_path = input("Enter the path of the proxy list file: ").strip()
    save_path = input("Enter the path to save the alive proxies: ").strip()

    # 1. Setup save directory and ensure the save file is clean
    if os.path.dirname(save_path) and not os.path.exists(os.path.dirname(save_path)):
        os.makedirs(os.path.dirname(save_path))

    # Clear the target file before starting to ensure a fresh list
    with open(save_path, 'w') as file:
        file.write("")

    proxies = load_proxies(file_path)
    if not proxies:
        return

    print(f"Starting check for {len(proxies)} proxies. Target: {PROXY_LIMIT}")

    # 2. Use ThreadPoolExecutor to run checks concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=50) as executor: # Increased workers for faster initial results
        future_to_proxy = {executor.submit(check_proxy, proxy): proxy for proxy in proxies}

        # 3. Process results as they complete
        for future in concurrent.futures.as_completed(future_to_proxy):
            proxy, is_alive = future.result()

            # Check if the overall process should stop
            if alive_count >= PROXY_LIMIT:
                # Cancel any remaining pending futures
                for future_to_cancel in future_to_proxy:
                    future_to_cancel.cancel()
                print("\nTarget limit reached. Stopping further checks...")
                break # Exit the as_completed loop

            if is_alive:
                # 4. Immediately write the proxy to the file and check the limit
                successfully_wrote = write_alive_proxy(proxy, save_path)

                if not successfully_wrote: # Should not happen, but safe check
                    continue
            else:
                # Only print "dead" if we haven't hit the limit yet to avoid clutter
                if alive_count < PROXY_LIMIT:
                    print(f"Proxy {proxy} is dead")

    print(f"\n--- Process Complete ---")
    print(f"Found {alive_count} alive proxies.")
    print(f"Alive proxies saved to {save_path}")

if __name__ == "__main__":
    main()
```
